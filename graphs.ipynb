```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
blip = [13.9, 31.2, 57.4]
ofa = [10.2, 25.1, 51.9]
git = [12.9, 32.8, 57.9]
qwen = [21.7, 42.1, 61.2]

small = [13.9, 10.2, 12.9, 21.7]
medium = [31.2, 25.1, 32.8, 42.1]
large = [57.4, 51.9, 57.9, 61.2]
```

```python
captions_models = ["BLIP-2", "OFA", "GIT", "Qwen2-VL"]
types = ["Small", "Medium", "Large"]
width = 0.2
x = np.arange(len(types))
print(x)
plt.figure(figsize = (11, 5))
bars = plt.bar(x = x - 1.5 * width, height = blip, width = width, label = "BLIP-2 + DINO")
axes = plt.gca()
axes.bar_label(bars)

bars = plt.bar(x = x - 0.5 * width, height = ofa, width = width, label = "OFA + DINO")
axes = plt.gca()
axes.bar_label(bars)

bars = plt.bar(x = x + 0.5 * width, height = git, width = width, label = "GIT + DINO")
axes = plt.gca()
axes.bar_label(bars)

bars = plt.bar(x = x + 1.5 * width, height = qwen, width = width, label = "Qwen2-VL + DINO")
axes = plt.gca()
axes.bar_label(bars)
plt.xticks(x, types)

plt.legend()

plt.title("Captioning-DINO Model AP By Object Area")
plt.ylabel("AP (%)")
plt.xlabel("Object Area")
```
